```python
from gtts import gTTS
from IPython.display import Audio
import random
import os
import time

def speak(words):
  tts = gTTS(words)
  sound_file = '1.wav'
  tts.save(sound_file)
  display(Audio(sound_file, autoplay=True))
  time = len(words)/5
  os.system('sleep ' + str(time))
  #time.sleep(2.2)


#speak("Hi, how are you boy")
#speak("what is your name")
#speak("Hello")

words_weeks_order = [
                        ["strong", "wrong", "copy", "cloth", "problem"],
                        ["strong", "wrong", "copy", "cloth", "problem"],
                        ["such", "much", "young", "unit", "united", "hue", "music", "human"],
                    ]
speak("which week test do you want to try")
week = input("which week test do you want to try " + "(1-" + str(len(words_weeks_order)) + "):")
week = int(week)
words_order = words_weeks_order[week-1]
length = len(words_order)
print(length)

speak("start test")

os.system('sleep 1')

words = random.sample(words_order, len(words_order))

for number in range(0, length):
    word = words[number]
    while(1):
        os.system('sleep 1.5')
        speak(word)
        i = input("spell:")

        if i == words[number]:
            print("--------------right-----------------")
            if number != length-1:
                speak("right next")
            break
        else:
            print("!!!!!!!!!!!!!!!!wrong try again")
            speak("wrong try again")

speak("hurry you pass this test")
```
